```python
def search_string_in_files(directory, search_string):
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                if file.endswith('.pth'):
                    data = torch.load(file_path)
                    print (f'successfully loaded file_path: {file_path}')
                    if search_string in str(data):
                        print(f"Found in: {file_path}")
                else:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        if search_string in f.read():
                            print(f"Found in: {file_path}")
            except Exception as e:
                print(f"Could not read {file_path}: {e}")
```

```python
directory = zenodo_dir
search_string = '00673f64-bad2-11e8-b2b8-ac1f6b6435d0'
search_string_in_files(directory, search_string)
```
